```python
import pandas as pd
data = pd.read_csv("Restaurant_Reviews.tsv",sep='\t')
#pd.read_table
```

```python
print(data)
```

```python
#Importing the packages
from nltk.tokenize import WhitespaceTokenizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import re

stop_words = stopwords.words("english")
tokenizer = WhitespaceTokenizer()
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
```

```python
corpus = []
for i in range(len(data.Review)):
    text = data.Review[i].lower()
    text = re.sub('[^a-z0-9]', ' ', text)
    tokens =  tokenizer.tokenize(text)
    filtered_tokens = [w for w in tokens if w not in stop_words]
    filtered = [stemmer.stem(lemmatizer.lemmatize(words)) for words in filtered_tokens]
    filtered_text = ' '.join(filtered)
    corpus.append(filtered_text)
```

```python
corpus[:5]
```

```python
#Bag of Words (BoW)technique to convert corpus into X
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
cv.fit(corpus)
```

```python
cv.vocabulary_.keys()
```

```python
features = list(cv.vocabulary_.keys())
features = sorted(features)
```

```python
cv.vocabulary_
```

```python
X =  pd.DataFrame(cv.transform(corpus).toarray(),columns= features)
y = data.Liked
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
cross_val_score(LogisticRegression(),X,y,cv=4).mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier()
model2.fit(X_train,y_train)
model2.score(X_test,y_test)
```
